```python
import pandas as pd
import numpy as np
```

```python
frame = pd.read_csv("/content/strdata.txt", sep="\s+")
frame
```

1. Fassen Sie Namen und Nachnamen in einer einzigen Spalte zusammen.

```python
frame["Fullname"] = frame["Name"] + " " + frame["Surname"]
frame = frame.drop(columns = ["Name"])
frame = frame.drop(columns = ["Surname"])
frame
```

2. Konvertieren Sie "Land" in einen kategorischen Datentyp und kodieren Sie dann die LÃ¤nder als Ganzzahlen (z. B. Telefonvorwahlen).


```python
frame["Country"] = frame['Country'].astype('category')
frame["Country"]
```

```python
frame["Country"] = frame["Country"].cat.codes
frame
```

```python
d = {0: "+49", 1 : "+90", 2 : "+44"}
deneme = pd.Series(d)

frame["Country-Code"] = np.where(frame["Country"].isin(deneme.index), deneme[frame["Country"]], frame.Country)  # BAÅARDIIIIIIIIIM
frame
```

3. Zerlegen Sie die Telefonnummern in Ortsvorwahl und Rufnummer.

```python
frame["Tel"] = frame['Tel'].astype('string')
frame["Tel"]
```

```python
frame["No"] = frame["Country-Code"] + frame["Tel"]
frame["Area-Code"] = frame["Tel"].str.slice(start=0, stop=3, step=1)
frame["Area-Code"] = frame["Tel"].str.slice(start=3, stop=10, step=1)
frame
```

4. Aufteilen (mit Regexp) von Benutzernamen, DomÃ¤nennamen und Suffixen in separate Spalten.

```python
import re
# pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
pattern = r"([A-Z0-9._%+-]+)@([A-Z]{2,4})\.([A-Z0-9.-]+)"
regex = re.compile(pattern, flags=re.IGNORECASE)
```

```python
b_array = []
d_array = []
s_array = []
for i in frame["e-mail"]:
  m = regex.match(i)
  m = m.groups()
  b_array.append(m[0])
  d_array.append(m[1])
  s_array.append(m[2])
```

```python
frame["benutzername"] = b_array
frame["domane"] = d_array
frame["suffix"] = s_array
```
