```python
import seaborn as sns
import joblib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = joblib.load("dfo.joblib")
df
```

```python
fig, axes = plt.subplots(2, 3, figsize=(10,6))
fig.tight_layout(pad=3.0)
```

```python
axes[0,0].scatter(df.lon, df.lat, alpha=0.4, s=.2)
axes[0,0].set_xlabel("lon")
axes[0,0].set_ylabel("lat(deg)")

axes[0,1].scatter(df.year, df.lon, alpha=0.4, s=.2)
axes[0,1].set_xlabel("time(year)")
axes[0,1].set_ylabel("lon")

axes[0,2].scatter(df.year, df.lat, alpha=0.4, s=.2)
axes[0,2].set_xlabel("time(year)")
axes[0,2].set_ylabel("lat(deg)")

axes[1,0].hist(df.lat, 50, alpha=0.4)
axes[1,0].set_xlabel("lat(deg)")
axes[1,0].set_ylabel("frequency")

axes[1,1].hist(df.carea, 50, log=True, alpha=0.4)
axes[1,1].set_xlabel("carea")
axes[1,1].set_ylabel("frequency")

axes[1,2].scatter(df.year, df.carea, alpha=0.4, s=.2)
axes[1,2].set_xlabel("time(year)")
axes[1,2].set_ylabel("carea")

fig
```

```python
fig = plt.figure(figsize=(8,10))
p = sns.scatterplot(data = df[df.year < 1962], x ="lon", y = "year", size ="carea")
```

### sns.kdeplot with bw_adjust=.5, alpha=.5, thresh=.5


```python
fig = plt.figure(figsize=(8,10))
p = sns.scatterplot(data = df[df.year < 1962], x ="lon", y = "year", size ="carea", sizes=(20,200), alpha=.3,
                    hue="lat", palette="dark:salmon_r")

sns.kdeplot(data = df[df.year < 1962], x ="lon", y = "year", bw_adjust=.5, alpha=.5, thresh=.5, color="blue")
```
